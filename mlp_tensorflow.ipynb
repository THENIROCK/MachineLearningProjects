```python
import numpy as np
from random import random
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

create a multi-layer perceptron with tensorflow.

```python
# i want to add two numbers and make it think they add to 42 every time
# build model, compile model, train model, evaluate model, make predictions

def generate_dataset(num_samples, test_size):
  # inputs
  x = np.array([[random()/2 for _ in range(2)] for _ in range(num_samples)])
  # outputs
  y = np.array([[(i[0] + i[1])/100000000000 + 0.42] for i in x])
  # training set
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size)
  return x_train, x_test, y_train, y_test

if __name__ == "__main__":
  #5000 samples, save 30% for test set
  x_train, x_test, y_train, y_test = generate_dataset(5000, 0.3)
  # print(x_test[0])
  # print(y_test[0])

  # build model
  model = tf.keras.Sequential([
      # 5 neurons, 2 dim input (2 numbers)
      tf.keras.layers.Dense(5, input_dim=2, activation="sigmoid"),
      tf.keras.layers.Dense(1, activation="sigmoid")
  ]) # left to right ML model

  # compile model
  optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)
  model.compile(optimizer=optimizer, loss="MSE")

  # train model
  model.fit(x_train, y_train, epochs=100) #LOL

  # evaluate model
  model.evaluate(x_test, y_test, verbose=1)

  # make predictions
  data = np.array([[0.1,0.2],[0.3,0.7]])
  predictions = model.predict(data)
  print("\nPredictions:")
  for data, prediction in zip(data,predictions):
    print("{} + {} = {}".format(data[0],data[1],prediction[0]*100))
```
